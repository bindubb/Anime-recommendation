```python
#let us try droping some unwanted columns and retaining only the usefull ones to build the recommendation systme
animes = animes[['id', 'title', 'genres', 'synopsis', 'image_url']]
animes
```

```python
animes['tags'] = animes['genres'] + animes['synopsis']
animes
```

```python
new_data = animes.drop(columns=['genres', 'synopsis'])
new_data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 10000, stop_words = 'english')
cv
```

```python
vectors = cv.fit_transform(animes['tags'].values.astype('U')).toarray()
vectors
```

```python
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
similarity
```

```python
similarity[2]
```

```python
dis = sorted(list(enumerate(similarity[2])), reverse = True, key = lambda vector:vector[1])
for i in dis[0:5]:
    print(new_data.iloc[i[0]].title)
```

```python
def recommand(animes):
    index = new_data[new_data['title']==animes].index[0]
    dis = sorted(list(enumerate(similarity[index])), reverse = True, key = lambda vector:vector[1])
    for i in dis[0:10]:
        print(new_data.iloc[i[0]].title)
       

        
```

```python
recommand("naruto")
```
